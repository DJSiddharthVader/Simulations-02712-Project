# Small Model Demo
Using the functions defined in `model.py` and `matrix.py` we simulate the model described in Eldar 2011 for 2 strains.

```python
import matrix as mx
import model as md

# make matrix
K_ac = mx.pattern_matrix('ident', 2)
print('QS interaction matrix')
print(K_ac)

# initial strain abundances
initial_N = (0.0001, 0.0002)
time_steps = 200

print('simulating...')
t, N, S, E, P = md.simulate(K_ac, initial_N, time_steps, md.params)
print('finished simulating!')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(t,N[0],label="N1")
plt.plot(t,N[1],label="N2")
plt.plot(t,S[0],label = "S1")
plt.plot(t,S[1],label = "S2")
plt.legend()
plt.show()
```
