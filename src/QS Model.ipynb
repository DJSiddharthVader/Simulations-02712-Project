## Model Demo
Using the functions defined in `model.py` and `matrix.py` we simulate the model described in Eldar 2011 for 2 strains.

### Set up Conditions for the simulation

```python
# initial strain abundances
initial_N = (0.00015, 0.002, 0.001, 0.0005)
time_steps = 200
```

### Simulate the null model

```python
K_ac = matrix.pattern_matrix('null', len(initial_N))
analysis.run_model(K_ac, initial_N, time_steps)
```

### Simulate faculative cheater model

```python
K_ac = matrix.pattern_matrix('ident', len(initial_N))
analysis.run_model(K_ac, initial_N, time_steps)
```
